```python
# 📷 Advanced Image Editor in Python (Fully Customizable)

!pip install Pillow

from PIL import Image, ImageEnhance, ImageFilter
from google.colab import files
from IPython.display import display

# Upload Image
print("📤 Upload your image (JPG/PNG)...")
uploaded = files.upload()
image_name = list(uploaded.keys())[0]
image = Image.open(image_name)
print("\n✅ Original Image:")
display(image)

# ==== CUSTOM EDIT OPTIONS ====

# Brightness
brightness_factor = float(input("💡 Enter brightness level (1=Normal, >1=Brighter, <1=Darker): "))
brightness = ImageEnhance.Brightness(image)
bright_image = brightness.enhance(brightness_factor)

# Contrast
contrast_factor = float(input("🎨 Enter contrast level (1=Normal, >1=High Contrast, <1=Low Contrast): "))
contrast = ImageEnhance.Contrast(bright_image)
contrast_image = contrast.enhance(contrast_factor)

# Sharpness
sharpness_factor = float(input("🔪 Enter sharpness level (1=Normal, >1=Sharper, <1=Blurry): "))
sharpness = ImageEnhance.Sharpness(contrast_image)
sharp_image = sharpness.enhance(sharpness_factor)

# Color Enhancement
color_factor = float(input("🌈 Enter color level (1=Normal, >1=More Vibrant, <1=Less Color): "))
color = ImageEnhance.Color(sharp_image)
color_image = color.enhance(color_factor)

# Blur
blur_amount = int(input("💨 Enter blur radius (0 = No Blur): "))
blurred_image = color_image.filter(ImageFilter.GaussianBlur(blur_amount))

# Resize (Optional)
resize_choice = input("📏 Do you want to resize the image? (y/n): ").lower()
if resize_choice == 'y':
    new_width = int(input("Enter new width: "))
    new_height = int(input("Enter new height: "))
    resized_image = blurred_image.resize((new_width, new_height))
else:
    resized_image = blurred_image

# Rotate (Optional)
rotate_choice = input("🔄 Do you want to rotate the image? (y/n): ").lower()
if rotate_choice == 'y':
    angle = int(input("Enter rotation angle (in degrees): "))
    final_image = resized_image.rotate(angle)
else:
    final_image = resized_image

# Convert to RGB before saving
final_image = final_image.convert("RGB")

# Display Final Image
print("\n✨ Final Edited Image:")
display(final_image)

# Save the Image
format_choice = input("💾 Save as JPG or PNG? (jpg/png): ").lower()
filename = input("Enter file name to save (without extension): ")

if format_choice == 'png':
    final_image.save(f"{filename}.png")
else:
    final_image.save(f"{filename}.jpg")

print("\n✅ Image saved successfully!")
```
